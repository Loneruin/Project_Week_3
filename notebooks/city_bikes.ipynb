# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import os
import pandas as pd
from IPython.display import JSON 
```

```python
url = "http://api.citybik.es/v2/networks/citi-bike-nyc"
headers = {'accept': 'application/JSON'}
response = requests.get(url, headers = headers)
```

```python
if response.status_code != 200:
    print(f'Error: {response.status_code}')
    print(response.text)
else:
    print(response)
```

```python
city_bike = response.json()
```

```python
JSON(city_bike)
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
stations = city_bike['network']['stations']
```

```python
stations
```

```python
for station in stations:
    latitude = station["latitude"]
    longitude = station["longitude"]
    bikes = station["free_bikes"]
    print(f'latitude: {latitude}, longitude: {longitude}, bikes: {bikes}')
```

Put your parsed results into a DataFrame.

```python
station_dict = []
for station in stations:
    latitude = station["latitude"]
    longitude = station["longitude"]
    empty_slots = station["empty_slots"]
    bikes = station["free_bikes"]
    station_dict.append({"latitude": latitude, "longitude": longitude,"empty_slots": empty_slots, "bikes": bikes})

nyc_bike = pd.DataFrame(station_dict)
```

```python
nyc_bike
```
